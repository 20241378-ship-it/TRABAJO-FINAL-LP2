```python

# EXTRACCIÓN Y CONSOLIDACIÓN DE DATOS (JOBICY)


import requests
import pandas as pd
import time


# Consumo de la API de Jobicy para obtener ofertas reales
#   Garantiza un scraping no intrusivo y respeta los términos del servicio

def obtener_empleos_reales():
    url = "https://jobicy.com/api/v2/remote-jobs"
    
    
    cabeceras = {
        "User-Agent": "proyecto_scrapping/1.0"
    }
    
    # Parámetros moderados de búsqueda para no saturar la red y evitar que nos bloqueen por exceso de peticiones
    parametros = {
        "count": 50,        
        "industry": "dev"   
    }
    
    try:
        print("Accediendo a la API de Jobicy...")
        respuesta = requests.get(url, headers=cabeceras, params=parametros, timeout=10)
        
        if respuesta.status_code == 200:
            datos = respuesta.json()
            
            return datos.get("jobs", [])
        else:
            print(f"Error al acceder a Jobicy. Código de estado: {respuesta.status_code}")
            return []
            
    except requests.exceptions.RequestException as e:
        print(f"Error de conexión en Fuente 1: {e}")
        return []

def procesar_y_guardar_csv(empleos):
    """
    Identifica las variables necesarias, integra la información en un DataFrame
    de Pandas y almacena el resultado directamente en un archivo CSV limpio.
    """
    if not empleos:
        print("No se encontraron datos para procesar.")
        return

    dataset_final = []
    
    # Extracción de las variables clave para cada oferta de trabajo
    for job in empleos:
        datos_empleo = {
            "id": job.get("id"),
            "titulo": job.get("jobTitle"),
            "compania": job.get("companyName"),
            "descripcion": job.get("jobDescription"),
            "url_postulacion": job.get("url"),
            "salario_min_anual": job.get("annualSalaryMin"),
            "fecha_extraccion": time.strftime("%Y-%m-%d %H:%M:%S")
        }
        dataset_final.append(datos_empleo)
    
  


    # CONVERSIÓN Y ALMACENAMIENTO DIRECTO EN CSV CON PANDAS

    
    # Transformamos nuestra lista de diccionarios en un DataFrame 
    df_empleos = pd.DataFrame(dataset_final)
    
   
    archivo_salida = "reporte_empleos_jobicy.csv"
    
    # Exportamos a CSV 
    
    df_empleos.to_csv(archivo_salida, index=False, encoding="utf-8")
        
   
    print("           PROCESAMIENTO DE DATOS EXITOSO")

    print(f" Se procesaron y guardaron {len(df_empleos)} ofertas en '{archivo_salida}'.")

    # Mostrar una vista previa de las primeras filas y columnas en la consola del vsc para asegurar el correcto funcionaminto
    print("Vista previa del DataFrame generado (primeras 3 filas):")
    print(df_empleos[["id", "titulo", "compania", "salario_min_anual"]].head(3))

if __name__ == "__main__":

    ofertas = obtener_empleos_reales()
    procesar_y_guardar_csv(ofertas)
```
