```python


import requests
import pandas as pd
import time
from bs4 import BeautifulSoup

def obtener_empleos_reales():
    url = "https://jobicy.com/api/v2/remote-jobs"
    
    # Cabeceras éticas e identificables para que el servidor conozca nuestra identidad y proposito
    cabeceras = {
        "User-Agent": "proyecto_Unalm/1.0"
    }
    
    # Parámetros controlados para descargar una cantidad moderada de datos y no ser bloqueados por la pagina por sobresaturarla de peticiones
    parametros = {
        "count": 50,       
        "industry": "dev"   
    }
    
    try:
        print("Accediendo a la API de Jobicy...")
        # Realizamos la petición HTTP GET con un tiempo límite de espera de 10 segundos
        respuesta = requests.get(url, headers=cabeceras, params=parametros, timeout=10)
        
        if respuesta.status_code == 200:
            datos = respuesta.json()
            
            return datos.get("jobs", [])
        else:
            print(f"Error al acceder a Jobicy. Código de estado: {respuesta.status_code}")
            return []
            
    except requests.exceptions.RequestException as e:
        print(f"Error de conexión en la Fuente de la API: {e}")
        return []


# Identifica las variables necesarias, limpia el formato HTML de la descripción,
def procesar_y_guardar_csv(empleos):
   
    if not empleos:
        print("No se encontraron datos para procesar.")
        return

    dataset_final = []
    
    print("Iniciando el procesamiento y la limpieza de etiquetas HTML...")
    # Iteración sobre cada oferta para extraer y limpiar las variables clave
    for job in empleos:
        
        descripcion_raw = job.get("jobDescription", "")
        
        
        if descripcion_raw:
            descripcion_limpia = BeautifulSoup(descripcion_raw, "html.parser").get_text().strip()
        else:
            descripcion_limpia = ""

        # 3. Mapeo estructurado de las variables del dataset
        datos_empleo = {
            "id": job.get("id"),
            "titulo": job.get("jobTitle"),
            "compania": job.get("companyName"),
            "descripcion": descripcion_limpia,  
            "url_postulacion": job.get("url"),
            "salario_min_anual": job.get("annualSalaryMin"),
            "fecha_extraccion": time.strftime("%Y-%m-%d %H:%M:%S") # Marca de tiempo por lotes
        }
        dataset_final.append(datos_empleo)
    
    
    # INTEGRACIÓN Y ESTRUCTURACIÓN DE TABLAS CON PANDAS
   
    
    # Convertimos nuestra lista de diccionarios en un DataFrame de Pandas
    df_empleos = pd.DataFrame(dataset_final)
    
    # Definimos el nombre del archivo CSV
    archivo_salida = "reporte_empleos_jobicy.csv"
    
   
    df_empleos.to_csv(archivo_salida, index=False, encoding="utf-8")
        

    print("           PROCESAMIENTO Y LIMPIEZA EXITOSA")
   
    print(f" Se procesaron, limpiaron y guardaron {len(df_empleos)} ofertas en '{archivo_salida}'.")
    print(" Estructura tabular generada de forma limpia y transparente.")
    print("=======================================================\n")
    
    # Mostrar una vista previa de los datos limpios en la consola para verificación del correcto funciomiento del codigo
    print("Vista previa del DataFrame consolidado (primeras 3 filas):")
    print(df_empleos[["id", "titulo", "compania", "salario_min_anual"]].head(3))

if __name__ == "__main__":

    ofertas_extraidas = obtener_empleos_reales()
    procesar_y_guardar_csv(ofertas_extraidas)
```
